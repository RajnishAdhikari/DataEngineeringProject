#  With date time , now creating a new column month, year , quarter

```python
#now converting the type in FL_DATE in datetime format 
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])

# createting new columns for year, month, and quarter
df['Year'] = df['FL_DATE'].dt.year
df['Quarter'] = df['FL_DATE'].dt.quarter
df['Month'] = df['FL_DATE'].dt.month

```

# creating new table to directly load the csv file data


```python
# Creating the   table
cursor = mydb.cursor()
cursor.execute("CREATE TABLE IF NOT EXISTS flight_data (FL_DATE DATE, OP_CARRIER VARCHAR(255), OP_CARRIER_FL_NUM INTEGER, ORIGIN VARCHAR(255), DEST VARCHAR(255), CRS_DEP_TIME INTEGER, DEP_TIME FLOAT, DEP_DELAY FLOAT, TAXI_OUT FLOAT,WHEELS_OFF FLOAT,WHEELS_ON FLOAT,TAXI_IN FLOAT,CRS_ARR_TIME INTEGER,ARR_TIME FLOAT,ARR_DELAY FLOAT,CANCELLED FLOAT,DIVERTED FLOAT, CRS_ELAPSED_TIME FLOAT,ACTUAL_ELAPSED_TIME  FLOAT,AIR_TIME FLOAT,DISTANCE FLOAT,Year INTEGER, Quarter INTEGER, Month INTEGER)");
```

```python
final_df = pd.read_csv('finaldata.csv')
final_df = final_df.replace(np.nan, 0)
insert_query = "INSERT INTO flight_data (FL_DATE,OP_CARRIER, OP_CARRIER_FL_NUM, ORIGIN, DEST,CRS_DEP_TIME,DEP_TIME,DEP_DELAY,TAXI_OUT,WHEELS_OFF,WHEELS_ON,TAXI_IN,CRS_ARR_TIME,ARR_TIME,ARR_DELAY,CANCELLED,DIVERTED,CRS_ELAPSED_TIME,ACTUAL_ELAPSED_TIME,AIR_TIME,DISTANCE,Year,Quarter,Month) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s, %s, %s, %s)"
values = final_df.values.tolist()
cursor.executemany(insert_query, values)
mydb.commit()
```

# Creating another table and load that table with data from previous created table.

```python
# create engine to connect to MySQL database
engine = sqlalchemy.create_engine('mysql+mysqlconnector://root:root@127.0.0.1:3307/aviationdata')

# create connection object
connection = engine.connect()

# create new table
connection.execute('CREATE TABLE IF NOT EXISTS new_flight_data LIKE flight_data')

# insert data from the previous table into the new table
connection.execute('INSERT INTO new_flight_data SELECT * FROM flight_data')

# close the connection
connection.close()

```
